```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

dataMapping = {
            "Y" : 1,
            "N" : 0
}

df1["Cancer"] = df1["Cancer"].map(dataMapping)
df1
```

```python
data = df1.copy()
```

```python


plt.figure(figsize=(10, 5))  # Adjust the figure size as needed

# Subplot 1: Countplot
plt.subplot(1, 2, 1)
sb.countplot(data=data, x='cancer')
plt.title('Cancer Value Counts')

# Subplot 2: Pie Chart
plt.subplot(1, 2, 2)
s = data['cancer'].value_counts()
labels = s.index
plt.pie(s, labels=labels, autopct='%.2f%%')
plt.title('Cancer Percentage Distribution')

# Remove plt.show() from the code, and add plt.tight_layout() for better spacing
plt.tight_layout()
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,5))
sb.heatmap(data.isnull(),cbar=True,cmap='viridis')
plt.show()
```
